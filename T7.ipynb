**Definição T7.1:** Seja $f:\mathcal S_1\to\mathcal S_2$ uma aplicação suave. Dado $p\in\mathcal S_1$ e $v\in T_p\mathcal S_1$, por definição existe $\gamma:I\to \mathcal S_1$ e $t_0\in I$ tal que $\gamma(t_0)=p$ e $\dot{\gamma}(t_0)=v$. Então $\widetilde{\gamma}=f\circ \gamma:I\to \mathcal S_2$ é uma curva paramétrica tal que $\widetilde{\gamma}(t_0)=f(p)$ e $w=\dot{\widetilde{\gamma}}(t_0)\in T_{f(p)}\mathcal S_2$. Através desta contrução, definimos *derivada de $f$ em $p$* como a aplicação $D_pf:T_p\mathcal S_1\to T_{f(p)}\mathcal S_2$ definida por $D_pf(v)=w$.

**Exercício:** Mostre que a definição T7.1 depende apenas de $f$, $p$ e $v$, ou seja, que não depende da escolha da curva $\gamma$.

**Proposição T7.2:** Seja  $f:\mathcal S_1\to\mathcal S_2$ uma aplicação suave e $p\in\mathcal S_1$. Então $D_pf$ é uma aplicação linear. Consideremos $\sigma$ e $\widetilde{\sigma}$ tais que:

1. $\sigma(t,s)$ é uma parametrização de um aberto de $\mathcal S_1$ que contém $p=\sigma(t_0,s_0)$,

2. $\widetilde{\sigma}(\widetilde{t},\widetilde{s})$ é uma parametrização de um aberto de $\mathcal S_2$ que contém $f(p)$,

3. $f(\sigma(t,s))=\widetilde{\sigma}(\alpha(t,s),\beta(t,s))$, com $\alpha,\beta$ suaves num aberto de $\mathbb R^2$.

A matriz de $D_pf$ relativamente às bases $\{\partial\sigma_t,\partial\sigma_s\}$ e $\{\partial\widetilde{\sigma}_{\widetilde{t}},\partial\widetilde{\sigma}_{\widetilde{s}}\}$, calculadas nos pontos adequados, é a matriz jacobiana da aplicação $(t,s)\mapsto (\alpha(t,s),\beta(t,s))$ no ponto $(t_0,s_0)$, ou seja,
$$
\begin{bmatrix}
\partial_t\alpha(t_0,s_0) && \partial_s\alpha(t_0,s_0)\\
\partial_t\beta(t_0,s_0) && \partial_s\beta(t_0,s_0)
\end{bmatrix}
$$

**Exemplo T7.3:** Sejam $a,b,c>0$.
$$
\mathcal E=\left\{(x_1,x_2,x_3)\in\mathbb R^3:\frac{x_1^2}{a^2}+\frac{x_2^2}{b^2}+\frac{x_3^2}{c^2}=1\right\}.
$$
Considere a aplicação $f:S^2\to \mathcal E$ definida por $f(x_1,x_2,x_3)=(ax_1,bx_2,cx_3)$. Calcule a derivada de $f$ nos pontos do aberto $\{(x_1,x_2,x_3)\in\mathbb R^3:x_1^2+x_2^2+x_3^2=1\wedge\,x_3>0\}$.

**Proposição T7.4:**

1. Dada uma superfície $\mathcal S$, a derivada da identidade em $\mathcal S$ é a identidade em $T_pS$.

2. Sejam $S_1,S_2,S_3$ superfícies. Dadas aplicações suaves $f:\mathcal S_1\to\mathcal S_2$ e $g:\mathcal S_2\to\mathcal S_3$ aplicações suaves, temos que

$$
\forall p\in\mathcal S_1, D_p(g\circ f)=D_{f(p)}g\circ D_pf.
$$

3. Se $f:\mathcal S_1\to\mathcal S_2$ é um difeomorfismo entre superfícies, então, para todo o $p\in\mathcal S_1$, $D_pf$ é um isomorfismo linear.

**Definição T7.5:** Dizemos que uma aplicação suave $f:\mathcal S_1\to\mathcal S_2$ entre superfícies é um *difeomorfismo local* se, para todo o $p\in\mathcal S_1$, existe um aberto $W$ de $\mathcal S_1$ que contém $p$ tal que $f(W)$ é um aberto de $\mathcal S_2$ e $f_{|W}:W\to f(W)$ é um difeomorfismo entre superfícies.

**Proposição T7.6:** Seja  $f:\mathcal S_1\to\mathcal S_2$ uma aplicação suave entre superfícies. Então $f$ é um difeomorfismo local se e só se, para todo o $p\in\mathcal S_1$, $D_pf$ é um isomorfismo.
