**Definição de curvatura (para uma curva suave regular):** Seja $\gamma:I\to \mathbb R^3$ uma curva suave regular. Definimos curvatura de $\gamma$ no ponto $\gamma(t)$, que denotamos por $\kappa(t)$, como a curvatura no ponto $\gamma(t)$ de uma sua reparametrização com velocidade unitária.

**Exercício T3.1:** Prove que a definição de curvatura não depende da escolha da reparametrização com velocidade unitária.

**Proposição T3.2:** Seja $\gamma:I\to \mathbb R^3$ uma curva suave regular. Então
$$
\kappa(t)=\frac{\|\ddot{\gamma}(t)\times \dot{\gamma}(t)\|}{\|\dot{\gamma}(t)\|^3}.
$$

**Exemplo T3.3:** Sejam $a,b\in\mathbb R$ e $I$ um intervalo aberto. Vejamos que a hélice $\gamma(t)=(a\cos(t),a\sin(t),bt)$, $t\in I$, é uma curva suave regular e calculemos a sua curvatura.

```python
from sympy import *
t, a, b = symbols('t a b')
dg=[diff(a*cos(t),t),diff(a*sin(t),t),diff(b*t,t)]
nonlinsolve(dg, [t])
```

```python
g=Matrix([a*cos(t),a*sin(t),b*t]).T
dg=g.diff(t)
dg
```

```python
d2g=dg.diff(t)
d2g
```

```python
c1= simplify(d2g.cross(dg))
c1
```

```python
simplify(sqrt(c1.dot(c1))/sqrt(dg.dot(dg))**3)
```

**Definição T3.4:** Seja $\gamma:I\to \mathbb R^2$ uma curva plana de classe $C^2$ com velocidade unitária. Seja $n_s(t)$ o vetor normal à curva, no ponto $\gamma(t)$, obtido aplicando uma rotação de $\pi/2$ a $\dot{\gamma}(t)$. Denomimanos $n_s(t)$ por *vetor normal unitário com sinal* de $\gamma$.

**Observação:** Dado $t\in I$, como $\gamma$ tem velocidade unitária, $\ddot{\gamma}(t)$ é perpendicular a $\dot{\gamma}(t)$, logo $n_s(t)$ e $\ddot{\gamma}(t)$ são linearmente dependentes, pelo que existe $\kappa_s(t)\in\mathbb R$ tal que
    $$
    \ddot{\gamma}(t)=\kappa_s(t) n_s(t).
    $$

**Definição de Curvatura com sinal (para uma curva plana com velociade unitária):** Seja $\gamma$ uma curva plana de classe $C^2$ com velocidade unitária. Denominamos $\kappa_s(t)$ por *curvatura com sinal* de $\gamma$ no ponto $\gamma(t)$.

**Proposição T3.5:** Seja $\gamma:I\to \mathbb R^2$ uma curva plana de classe $C^2$ com velocidade unitária. Consideremos $\gamma$ como uma curva de $\mathbb R^3$ contida no plano $\{(x,y,z)\in\mathbb R^3:z=0\}$. Então $\kappa=|\kappa_s|$.

**Proposição T3.6:** Seja $\gamma:I\to \mathbb R^2$ uma curva plana de velocidade unitária, $s_0\in I$ e $\varphi_0\in\mathbb R$ tal que 
    $$
    \dot{\gamma}(s_0)=(\cos(\varphi_0),\sin(\varphi_0)).
    $$
Então existe uma e uma só função suave $\varphi:I\to \mathbb R$ tal que $\varphi(s_0)=\varphi_0$ e, para todo o $s\in I$,
    $$
    \dot{\gamma}(s)=(\cos(\varphi),\sin(\varphi)).
    $$
