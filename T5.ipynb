**Definição T5.1:** Seja $A\subseteq \mathbb R^m$, $B\subseteq\mathbb R^n$ e $f:A\to B$ uma aplicação. Dado $a\in A$, dizemos que $f$ é *contínua em $a$* se
$$
\forall \epsilon>0, \exists \delta>0: x\in A\wedge \|x-a\|_{\mathbb R^m}<\delta \Rightarrow \|f(x)-f(a)\|_{\mathbb R^n}<\epsilon.
$$
Dizemos que $f$ é *contínua em $A$* se $f$ é contínua em todos os pontos de $A$.

**Nota:** A aplicação $f:A\to B$ é contínua em $A$ se e só se para todo o conjunto aberto $V$ de $\mathbb R^n$, existe um conjunto aberto $U$ de $\mathbb R^n$ tal que
$$
U\cap A=f^{-1}(V\cap B).
$$

**Exemplo T5.2:** Seja $S^2=\{(x,y,z)\in\mathbb R^3:x^2+y^2+z^2=1\}$. A aplicação $f:[0,2\pi]\times[0,\pi]\to S^2$ definida por
$$
f(\theta,\varphi)=(\sin(\varphi)\cos(\theta),\sin(\varphi)\sin(\theta),\cos(\varphi))
$$
é uma aplicação contínua.

**Definição T5.4:** Seja $A\subseteq \mathbb R^m$, $B\subseteq\mathbb R^n$ e $f:A\to B$ uma bijeção. Dizemos que $f$ é um *homeomorfismo* se $f$ e $f^{-1}$ são contínuas em $A$ e $B$, respectivamente. Nestas condições dizemos que $A$ e $B$ são homeomorfos.

**Definição T5.6:** Seja $A\subseteq \mathbb R^n$. Dizemos que $W\subseteq  A$ é um *conjunto aberto de $A$* se existe um conjunto aberto $V$ de $\mathbb R^n$ tal que $W=V\cap A$.

**Definição de superfície:** Seja $\mathcal S\subseteq \mathbb R^3$. Dizemos que $\mathcal S$ é uma *superfície* se, para todo o ponto $a\in \mathcal S$, existe um conjunto aberto $U$ de $\mathbb R^2$ e um conjunto aberto $W$ de $\mathcal S$ tais que $a\in W$ e os abertos $U$ e $W$ são homeomorfos. Denominamos por parametrização de $W$ ao homeomorphismo $\sigma:U\to W$. Dizemos que uma coleção $\{\sigma_{\alpha}:U_{\alpha}\to W_{\alpha}\}_{\alpha\in\mathcal I}$, de parametrizações de abertos de $\mathcal S$, é um atlas de $\mathcal S$ se 
$$
\cup_{\alpha\in\mathcal I}W_{\alpha}=\mathcal S.
$$

**Definição T5.8:** Seja $U$ um conjunto aberto de $\mathbb R^m$ e $f=(f_1,\ldots,f_n):U\to \mathbb R^n$ uma função.Dizemos que $f$ é *suave* se $f\in C^{\infty}(U)$, ou seja, se $f_1,\ldots,f_n$ admitem derivadas parciais de qualquer ordem contínuas em $U$.

**Nota:** Vamos abreviar a notação derivada parcial relativamente a uma variável $x$ por $\partial_x$. 

**Definição T5.9:** Seja $S$ uma superfície, $W$ um aberto de $S$ e $\sigma:U\to W$ uma parametrização. Dizemos que $\sigma$ é regular se, vista como função de $U\to \mathbb R^3$ é suave e $\partial_t\sigma\times \partial_s\sigma$ é diferente de zero em todos os pontos de $U$. Dizemos que $\mathcal S$ é uma superfície *suave* se, para todo o ponto $a\in \mathcal S$, existe $\sigma:U\to W$ uma parametrização regular tal que $a\in W$. Dizemos que uma coleção $\{\sigma_{\alpha}:U_{\alpha}\to W_{\alpha}\}_{\alpha\in\mathcal I}$ é um atlas *regular* de $\mathcal S$ se $\{\sigma_{\alpha}:U_{\alpha}\to W_{\alpha}\}_{\alpha\in\mathcal I}$ é um atlas de $\mathcal S$ composto por parametrizações regulares.

**Nota:** A condição $\partial_t\sigma\times \partial_s\sigma$ é diferente de zero em todos os pontos de $U$ é equivalente a $\partial_t\sigma,\partial_s\sigma$ são linearmente independentes em todos os pontos de $U$.

**Exemplo T5.10:**
1. A esfera $S^2$ é uma superfície suave.

```python
from sympy import *
t, s = symbols('t s')
g=Matrix([t,s,sqrt(1-t**2-s**2)]).T
dgt=g.diff(t)
dgt
```

```python
dgs=g.diff(s)
dgs
```

```python
c=simplify(dgt.cross(dgs))
c
```
