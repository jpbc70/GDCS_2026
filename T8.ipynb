**Definição T8.1:** Seja $\sigma$ uma parametrização de um aberto de uma superfície $\mathcal S$ e $p=\sigma(t_0,s_0)$. Segundo a Proposição T6.12 o vetor unitário
$$
N_\sigma(t_0,s_0)=\frac{\partial_t\sigma(t_0,s_0)\times\partial_s\sigma(t_0,s_0)}{\|\partial_t\sigma(t_0,s_0)\times\partial_s\sigma(t_0,s_0)\|}
$$
é normal a $T_p\mathcal S$. Denominamos $N_\sigma(t_0,s_0)$ por *vetor normal unitário standard* da parametrização $\sigma$ em $p$.

**Notas:** 

1. Todo o plano de $\mathbb R^3$ admite apenas dois vetores normais unitários.

2. Como visto na demonstração da Proposição T6.1, se $\widetilde{\sigma}$ é outra parametrização de um aberto da superfície $\mathcal S$ que contém $p$, temos a igualdade, nos pontos adequados,
$$
\partial_{\widetilde{t}}\widetilde{\sigma}\times\partial_{\widetilde{s}}\widetilde{\sigma}=\textrm{Det}(J_{\varphi})\partial_t\sigma\times\partial_s\sigma,
$$
o que implica, $N_{\widetilde{\sigma}}=N_\sigma$ ou $N_{\widetilde{\sigma}}=-N_\sigma$.


**Definição T8.2:** Dizemos que uma superfície $\mathcal S$ é *orientável* se existe um atlas $\mathcal A$ tal que, para todo a aplicação $\varphi$ de transição entre parametrizações do atlas, $\textrm{Det}(J_{\varphi})>0$ em todos os pontos onde $\varphi$ está definida.

**Proposição T8.3:** Uma superfície $\mathcal S$ é orientável se e só se existe uma aplicação suave $N:\mathcal S\to \mathbb R^3$ tal que, para todo o $p\in S$, $N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$.

**Definição T8.4:** Uma *superfície orientada* é uma superficie dotada de uma aplicação suave $N:\mathcal S\to \mathbb R^3$ tal que, para todo o $p\in S$, $N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$.

**Definição T8.7:** Seja $\mathcal S$ uma superfície e $p\in\mathcal S$. Denominamos por *primeira forma fundamental de $\mathcal S$ em $p$* à restrição do produto interno usual em $\mathbb R^3$ a $T_p S$.

**Notação:** Seja $\mathcal S$ uma superfície, $p\in\mathcal S$, $\sigma$ uma parametrização de um aberto de $\mathcal S$ que contém $p$ e $\sigma(t_0,s_0)=p$. Vamos representar a base dual de $\{\partial_t\sigma(t_0,s_0),\partial_s\sigma(t_0,s_0)\}$ por $\{dt,ds\}$. Por definição, temos que $dt:T_p\mathcal S\to \mathbb R$ e $ds:T_p\mathcal S\to \mathbb R$. Dado $v=\lambda\partial_t\sigma(t_0,s_0)+\mu \partial_s\sigma(t_0,s_0)$, $\lambda,\mu\in\mathbb R$, então $dt(v)=\lambda$ e $ds(v)=\mu$.

**Proposição T8.8:** Seja $\mathcal S$ uma superfície, $p\in\mathcal S$, $\sigma$ uma parametrização de um aberto de $\mathcal S$ que contém $p$ e $\sigma(t_0,s_0)=p$. Então
$$
v.v=E\,dt(v)^2+2F\,dt(v)ds(v)+G\,ds(v)^2,
$$
onde $E=\|\partial_t\sigma(t_0,s_0)\|^2$, $F=\partial_t\sigma(t_0,s_0).\partial_s\sigma(t_0,s_0)$ e $G=\|\partial_s\sigma(t_0,s_0)\|^2$.

**Definição T8.9:** Seja $\mathcal S$ uma superfície e $\sigma$ uma parametrização de um aberto de $\mathcal S$. Denominamos
$$
E\,dt^2+2F\,dtds+G\,ds^2,
$$
onde $E=\|\partial_t\sigma\|^2$, $F=\partial_t\sigma.\partial_s\sigma$ e $G=\|\partial_s\sigma\|^2$, por *primeira forma fundamental de $\sigma$*.

**Exercício T8.10:** Calcule a primeira forma fundamental das seguintes parametrizações:

1. (esfera unitária)
$$
\sigma(t,s)=(\sin(s)\cos(t),\sin(s)\sin(t),\cos(s)),\ (t,s)\in ]0,2\pi[\times]0,\pi[.
$$

```python
from sympy import *
t, s = symbols('t s')
g=Matrix([sin(s)*cos(t),sin(s)*sin(t),cos(s)]).T
jgt=g.diff(t)
jgt
```

```python
jgs=g.diff(s)
jgs
```

```python
E= trigsimp(jgt.dot(jgt))
E
```

```python
F= trigsimp(jgt.dot(jgs))
F
```

```python
G= trigsimp(jgs.dot(jgs))
G
```

**Resposta:** $\sin^2(s)\,dt^2+ds^2$.
