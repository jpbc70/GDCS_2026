**Definição T11.1:**  Seja $\mathcal S$ uma superfície orientada, $\gamma:I\to S$ uma curva e $\mathcal N:\mathcal S\to \mathbb R^3$ uma aplicação suave tal que, para todo o $p\in S$, $\mathcal N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$. Fixemos $t_0,t_1\in I$ e sejam $p=\gamma(t_0)$ e $q=\gamma(t_1)$. Dado $v_0\in T_p S$, resolvendo o sistema de equações diferenciais dadas na proposição T10.9 com a condição inicial $v_0$, obtemos o único campo vetorial $v$ paralelo ao longo de $\gamma$ tal que $v(t_0)=v_0$. Definimos a aplicação de *transporte paralelo de $p$ para $q$ ao longo de $\gamma$*, que tem por imagem de $v_0$ o elemento $\prod_{\gamma}^{pq}(v_0)=v(t_1)\in  T_q S$.

**Proposição T11.2:** Seja $\mathcal S$ uma superfície orientada, $\gamma:I\to S$ uma curva e $p,q\in \gamma(I)$. A aplicação de transporte paralelo de $p$ para $q$ ao longo de $\gamma$, $\prod_{\gamma}^{pq}:T_p S\to T_q S$, é uma aplicação linear que é uma isometria.

**Definição T11.3:** Seja $\mathcal S$ uma superfície orientada e $p\in \mathcal S$. Definimos:

* A *curvatura Gaussiana* de $\mathcal S$, $\mathcal K$, como o determinante de $\mathcal W_{p,S}$.

* A *curvatura média* de $\mathcal S$, $\mathcal H$, como metade do traço de $\mathcal W_{p,S}$.

**Nota:** O determinante e o traço de uma aplicação linear não dependem da escolha de base. A curvatura Gaussiana não depende da orientação e a curvatura média, relativamente à orientação, é única a menos de uma sinal (porquê).

**Definição T11.4:** Seja $\mathcal S$ uma superfície e $\sigma(t,s)$ uma parametrização de um aberto de $\mathcal S$, com primeira forma fundamental
$$
E\,dt^2+2F\,dt\,ds+G\,ds^2
$$
e segunda forma fundamental
$$
L\,dt^2+2M\,dt\,ds+N\,ds^2.
$$
Definimos as matrizes $\mathcal F_I$ e $\mathcal F_{II}$ como
$$
\mathcal F_I=
\begin{bmatrix}
E & F\\
F & G
\end{bmatrix}
\textrm{ e }
\mathcal F_{II}=
\begin{bmatrix}
L & M\\
M & N
\end{bmatrix}.
$$

**Proposição T11.5:** Seja $\sigma(t,s)$ uma parametrização de um aberto $V$ de uma superfície orientada $\mathcal S$, com primeira forma fundamental
$$
E\,dt^2+2F\,dt\,ds+G\,ds^2
$$
e segunda forma fundamental
$$
L\,dt^2+2M\,dt\,ds+N\,ds^2.
$$
Sendo $p$ um ponto de $V$, a matriz de $\mathcal W_{p,\mathcal S}$ relativamente à base $\{\partial_t\sigma,\partial_s\sigma\}$ é  $\mathcal F_I^{-1}\mathcal F_{II}$.

**Corolário T11.6:** Com a notação da proposição T11.5,
$$
\mathcal K=\frac{LN-M^2}{EG-F^2} \textrm{ e } \mathcal H=\frac{LG-2MF+NE}{2(EG-F^2)}.
$$

**Exemplos T11.7:**
* (cilindro) Calcule a curvatura Gaussiana e média em todos os pontos da parametrização
$$
\sigma(t,s)=(\cos t,\sin t,s),\, (t,s)\in ]0,2\pi[\times \mathbb R.
$$

```python
from sympy import *
t, s, = symbols('t s')
g=Matrix([cos(t),sin(t),s]).T
gt=g.diff(t)
gs=g.diff(s)
E= trigsimp(gt.dot(gt))
F= trigsimp(gt.dot(gs))
G= trigsimp(gs.dot(gs))
Matrix([E,F,G]).T
```

```python
nt=gt.cross(gs)
nus=simplify(nt/sqrt(nt.dot(nt)))
nus
```

```python
gtt=gt.diff(t)
gts=gt.diff(s)
gss=gs.diff(s)
L=trigsimp(gtt.dot(nus))
M=trigsimp(gts.dot(nus))
N=trigsimp(gss.dot(nus))
Matrix([L,M,N]).T
```

```python
K=simplify((L*N-M**2)/(E*G-F**2))
K
```

```python
H=(L*G-2*M*F+N*E)/(2*(E*G-F**2))
H
```

**Teorema T11.9:** Seja $\sigma:U\to W$ uma parametrização de um aberto $W$ de uma superfície orientada $\mathcal S$ e $\mathcal N$ a normal untitária standard. suponha que $\delta>0$ é tal que o círculo fechado
$$
\overline{D}_{\delta}=\{(t,s)\in\mathbb R^2:(t-t_0)^2+(s-s_0)^2\leq \delta^2\}
$$
está contido em $U$. Então o módulo da curvatura Gaussiana de $\mathcal S$ em $\sigma(t_0,s_0)$ é 
$$
\lim_{\delta\to 0}\frac{\iint_{\overline{D}_{\delta}} 
\left\|\partial_t\mathcal N\times \partial_s\mathcal N\right\|dt\,ds.}{\iint_{\overline{D}_{\delta}} 
\left\|\partial_t\sigma\times \partial_s\sigma\right\|dt\,ds}.
$$

**Definição T11.10:** Seja $V$ um espaço vetorial real de dimensão finita munido de um produto interno $.$ e $T:V\to V$ uma aplicação linear. Dizemos que $T$ é um operador *auto-adjunto* se
$$
T(u).w=u.T(w),
$$
para todo o $u,w\in V$.

**Proposição T11.11:** Seja $V$ um espaço vetorial real de dimensão $n$ munido de um produto interno $.$ e $T:V\to V$ um operador auto-adjunto. Então $V$ admite uma base $\{w_1,\ldots,w_n\}$ de vetores próprios tais que se $w_i,w_j$ são vetores próprios associados a valores próprios distintos então $w_i.w_j=0$.

**Nota:** Seja $\mathcal S$ uma superfície orientada e $p\in\mathcal S$. A aplicação $\mathcal W_{p,\mathcal S}$ é um operador auto-adjunto, e como tal existe,  uma base $\{\mathbf{t}_1,\mathbf{t}_2\}$ de $T_p\mathcal S$ tal que
$$
\mathcal W_{p,\mathcal S}(\mathbf{t}_1)=\kappa_1\mathbf{t}_1 \textrm{ e } \mathcal W_{p,\mathcal S}(\mathbf{t}_2)=\kappa_2\mathbf{t}_2.
$$
Além disso, se $\kappa_1\neq \kappa_2$ então  $\mathbf{t}_1.\mathbf{t}_2=0$.

**Definição T11.12:** Denominamos os escalares $\kappa_1,\kappa_2$ da nota anterior como as *curvaturas principais* de $\mathcal S$ em $p$ e os vetores $\mathbf{t}_1,\mathbf{t}_2$ por *vetores principais* de $\kappa_1$ e $\kappa_2$, respetivamente. 

**Definição T11.13:** Seja $\mathcal S$ uma superfície orientada e $p\in\mathcal S$. Dizemos que $p$ é um *ponto umbilical* se $\kappa_1=\kappa_2$.

**Nota:** Um ponto é umbilical se e só se existe um escalar real $\alpha$ tal que  $\mathcal W_{p,\mathcal S}=\alpha I_{T_p\mathcal S}$, onde $I_{T_p\mathcal S}$ é a aplicação identidade em $T_p\mathcal S$.

**Proposição T11.14:** Seja $\mathcal S$ uma superfície orientada e $p\in\mathcal S$. Então $T_p\mathcal S$ admite uma base ortonormada constituída por vetores principais.

**Proposição T11.15:** Seja $\mathcal S$ uma superfície orientada e $p\in\mathcal S$. Então, em $p$, temos
$$
\mathcal K=\kappa_1\kappa_2 \textrm{ e } \mathcal H=\frac{1}{2}(\kappa_1+\kappa_2).
$$

**Teorema de Euler:** Seja $\mathcal S$ uma superfície orientada, $p\in\mathcal S$ e $\gamma:I\to \mathcal S$ uma curva de velocidade unitária que passa por $p$. Então
$$
\kappa_n=\kappa_1\textrm{cos}^2\theta+\kappa_2\textrm{sen}^2\theta,
$$
onde $\{\mathbf{t}_1,\mathbf{t}_2\}$ é uma base ortonormada de $T_p\mathcal S$ constituída por vetores principais e $\theta$ é o ângulo entre os vetores $\mathbf{t}_1$ e $\dot{\gamma}$.

**Corolário:** Seja $\mathcal S$ uma superfície orientada e $p\in\mathcal S$. As curvaturas principais de $\mathcal S$ correspondem ao valor máximo e mínimo da curvatura normal, percorrendo todas as curvas contidas em $\mathcal S$ e que passam em $p$. Os vetores principais são os vetores de $T_p\mathcal S$ que correspondem aos vetores tangentes em $p$ das curvas onde é obtido o máximo e o mínimo.

**Proposição T11.16:** Seja $\sigma(t,s)$ uma parametrização de um aberto $V$ de uma superfície orientada $\mathcal S$, com primeira forma fundamental
$$
E\,dt^2+2F\,dt\,ds+G\,ds^2
$$
e segunda forma fundamental
$$
L\,dt^2+2M\,dt\,ds+N\,ds^2.
$$
Sendo $p$ um ponto de $V$, relativamente à base $\{\partial_t\sigma,\partial_s \sigma\}$ de $T_p\mathcal S$, as curvaturas principais são as soluções da equação
$$
\begin{vmatrix}
L-\lambda E && M-\lambda F\\
M-\lambda F && N-\lambda G
\end{vmatrix}=0
$$
Para cada solução $\lambda$ do sistema acima indicado, os vetores principais têm por coordenadas as soluções do sistema homogéneo associado à matriz
$$
\begin{bmatrix}
L-\lambda E && M-\lambda F\\
M-\lambda F && N-\lambda G
\end{bmatrix}.
$$
