**Definição T12.1:** Seja $\mathcal S$ uma superfície orientada, $\gamma:I\to S$ uma curva. Dizemos que $\gamma$ é uma *geodésica* de $\mathcal S$ se, para todo o $t\in I$, $\nabla_{\gamma}\dot{\gamma}(t)=0$.

**Proposição T12.2:** Seja $\mathcal S$ uma superfície orientada e $\gamma:I\to \mathcal S$ uma geodésica de $\mathcal S$. Então $\|\dot{\gamma}(t)\|$ é constante. 

**Nota:** Uma reparametrização de uma geodésica não tem de ser uma geodésica, mas uma reparametrização de velocidade unitária de uma geodésica ainda é uma geodésica.

**Proposição T12.3:** Seja $\mathcal S$ uma superfície orientada e $\gamma:I\to S$ uma curva de velocidade unitária. Então $\gamma$ é uma geodésica se e só se $\kappa_g$ é nula em todos os pontos de $\gamma$.

**Proposição T12.5:** Seja $\mathcal S$ uma superfície orientada e $\gamma:I\to \mathcal S$ uma curva. A curva $\gamma$ é uma geodésica se e só se,para todo o arco de curva $\gamma(t)=\sigma(u(t),w(t))$ contido numa parametrização $\sigma(u,w)$ de um aberto de $\mathcal S$, as *equações geodésicas* são verificadas: 
$$
\frac{d}{dt}(E\dot{u}+F\dot{w})=\frac{1}{2}\left(\dot{u}^2\partial_u E+2\dot{u}\dot{w}\partial_u F+\dot{w}^2\partial_u G\right),
$$
$$
\frac{d}{dt}(F\dot{u}+G\dot{w})=\frac{1}{2}\left(\dot{u}^2\partial_w E+2\dot{u}\dot{w}\partial_w F+\dot{w}^2\partial_w G\right),
$$
onde $E\,dt^2+2F\,dt\,ds+G\,ds^2$ é a primeira forma fundamental de $\sigma$.

 **Proposição T12.6:** Seja $\mathcal S$ uma superfície orientada e $\gamma:I\to \mathcal S$ uma curva. A curva $\gamma$ é uma geodésica se e só se, para todo o arco de curva $\gamma(t)=\sigma(u(t),w(t))$ contido numa parametrização $\sigma(u,w)$ de um aberto de $\mathcal S$, as seguintes equações são verificadas:
 $$
\ddot{u}+\Gamma^{1}_{11}\dot{u}^2+2\Gamma^{1}_{12}\dot{u}\dot{w}+\Gamma^{1}_{22}\dot{w}^2=0,
$$
$$
\ddot{w}+\Gamma^{2}_{11}\dot{u}^2+2\Gamma^{2}_{12}\dot{u}\dot{w}+\Gamma^{2}_{22}\dot{w}^2=0.
$$

**Exemplo T12.7:** Calculemos as geodésicas da esfera $S^2=\{(x,y,z)\in\mathbb R^3:x^2+y^2+z^2=1\}$.

```python
from sympy import *
u, w, c= symbols('u w c')
g=Matrix([cos(u)*cos(w),cos(u)*sin(w),sin(u)]).T
gu=g.diff(u)
gw=g.diff(w)
E= trigsimp(gu.dot(gu))
F= trigsimp(gu.dot(gw))
G= trigsimp(gw.dot(gw))
FI=Matrix([[E,F],[F,G]])
FI
```

Obtemos o sistema de equações:
$$
\ddot{u}=-\frac{\sin(2u)}{2}\dot{w}^2,
$$
$$
\frac{d}{dt}(\cos^2(u)\dot{w})=0.
$$
Para simplificar, podemos considerar que a geodésica tem velocidade constante. Como a nossa parametrização tem primeira forma fundamental $du^2+cos^2(u)dw^2$ e $\dot{\gamma}=\dot{u}\partial_u\sigma+\dot{w}\partial_w\sigma$, então
$$
1=\dot{\gamma}.\dot{\gamma}=\dot{u}^2+\cos^2(u)\dot{w}^2.
$$
