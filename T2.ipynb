**Definição T2.1:** Seja $\gamma:I\to \mathbb R^n$ uma curva paramétrica, $t_0\in I$. Definimos a função *comprimento de arco* com início no ponto $\gamma(t_0)$ como
$$
s(t)=\int_{t_0}^t\| \dot{\gamma}(u)\| du,\, t\in I.
$$

**Exemplo T2.2:** Seja $k\in\mathbb R\setminus\{0\}$. Calcule a função comprimento de arco da curva paramétrica
$$
\gamma(t)=\left(e^{kt}\cos(t),e^{kt}\sin(t)\right),\, t\in\mathbb R,
$$
com início em $\gamma(0)=(1,0)$.

```python
from sympy import *
t, k, u = symbols('t k u')
plot_parametric((exp(0.1*t)*cos(t),exp(0.1*t)*sin(t)), (t, -9, 9))
```

```python
g=Matrix([[exp(k*t)*cos(t),exp(k*t)*sin(t)]])
g
```

```python
dg=g.diff(t)
dg
```

```python
ndg=sqrt(simplify(dg.dot(dg)))
ndg
```

```python
integrate(ndg.subs(t,u),(u,0,t))
```

**Nota:** A função comprimento de arco, para $t\geq t_0$ e supondo que $\gamma$ não é fechada, calcula o comprimento do arco de $\gamma$ que vai de $\gamma(t_0)$ até $\gamma(t)$. Quando a curva é fechada, a situação é mais delicada!

**Definição T2.3:** Uma curva paramétrica $\widetilde{\gamma}:\widetilde{I}\to \mathbb R^n$ é uma *reparametrização* de $\gamma:I\to \mathbb R^n$ se existe um difeomorfismo $\varphi:\widetilde{I}\to I$ tal que $\widetilde{\gamma}(t)=\gamma(\varphi(t)),\,\forall t\in \widetilde{I}$.

**Exercício T2.4:** A curva $\widetilde{\gamma}(t)=(\sin(t),\cos(t))$ é uma reparametrização de $\gamma(t)=(\cos(t),\sin(t)),\,t\in]-\pi,3\pi[$ pois?

**Proposição T2.5:** A reparametrização de uma curva regular é uma curva regular. 

**Proposição T2.6:** Uma curva paramétrica $\gamma$ admite uma reparametrização $\widetilde{\gamma}$ com velocidade unitária se e so só é regular. 

**Definição de Curvatura (para uma curva com velociade unitária):** Seja $\gamma$ uma curva de classe $C^2$ com velocidade unitária. Definimos *curvatura* de $\gamma$ no ponto $\gamma(t)$, que denotamos por $\kappa(t)$, como $\|\ddot{\gamma}(t)\|$.
