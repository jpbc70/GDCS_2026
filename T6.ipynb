**Notação:** Para facilitar a notação, identificamos uma parametrização $\sigma:U\to W$ de um aberto de uma superfície com $\sigma:U\to \mathbb R^3$, a extensão de $\sigma$ tendo como conjunto de chegada $\mathbb R^3$.

**Proposição T6.1:** Seja $\sigma:U\to \mathbb R^3$ uma parametrização regular e $\varphi:V\to U$ um difeomorphismo suave. Então $\widetilde{\sigma}=\sigma\circ\varphi:V\to\mathbb R^3$ é uma parametrização regular.

**Definição T6.2:** Sejam $\sigma:U\to \mathbb R^3$ e $\widetilde{\sigma}:V\to \mathbb R^3$ parametrizações regulares de abertos de uma mesma superfície. Se existir um difeomorfismo suave $\varphi:V\to U$ tal que $\widetilde{\sigma}=\sigma\circ\varphi$, dizemos que $\widetilde{\sigma}$ é uma *reparametrização* de $\sigma$.

**Definição T6.3:** Seja $\mathcal S$ uma superfície suave e $\{\sigma_{\alpha}:U_{\alpha}\to W_{\alpha}\}_{\alpha\in\mathcal I}$ um atlas regular de $\mathcal S$. Dizemos que $\tau_{\alpha,\beta}=\sigma^{-1}_{\beta}\circ\sigma_{\alpha}:\sigma^{-1}_{\alpha}(W_{\alpha}\cap W_{\beta})\to \sigma^{-1}_{\beta}(W_{\alpha}\cap W_{\beta})$, $\alpha,\beta\in \mathcal I$, é uma aplicação de transição se $W_{\alpha}\cap W_{\beta}\neq \emptyset$.

**Notas:** 
1. Pelo Teorema da Função Inversa, as aplicações de transição são difeomorfismos suaves (entre abertos de $\mathbb R^2$).
2. Dado que $(\sigma_{\alpha})_{|\sigma^{-1}_{\alpha}(W_{\alpha}\cap W_{\beta})}=(\sigma_{\beta})_{| \sigma^{-1}_{\beta}(W_{\alpha}\cap W_{\beta})}\circ \tau_{\alpha,\beta}$, $(\sigma_{\alpha})_{|\sigma^{-1}_{\alpha}(W_{\alpha}\cap W_{\beta})}$ é uma reparametrização de $(\sigma_{\beta})_{| \sigma^{-1}_{\beta}(W_{\alpha}\cap W_{\beta})}$.
