**Definição T9.1:** Seja $f:\mathcal S_1\to\mathcal S_2$ uma aplicação suave entre superfícies. Dizemos que $f$ é uma *isometria local* se para toda a curva suave $\gamma:I\to \mathcal S_1$ de comprimento $\ell(\gamma)$, a curva $f\circ \gamma$ tem comprimento $\ell(\gamma)$. Neste caso, dizemos que as superfícies $\mathcal S_1$ e $\mathcal S_2$ são *localmente isométricas*. 

**Notação:** Seja $f:\mathcal S_1\to\mathcal S_2$ uma aplicação suave entre superfícies e $p\in\mathcal S_1$. Para $v,w\in T_p\mathcal S_1$, definimos
$$
f^{\ast}(v.w)_p=D_pf(v).D_pf(w)\,.
$$


**Proposição T9.2:**  Para todo o $p\in\mathcal S_1$, $f^{\ast}( . )_p$ é uma forma bilinear simétrica, ou seja, para $\lambda,\mu\in\mathbb R$ e $u,v,w\in T_p\mathcal S_1$,
1. $f^{\ast}(v.w)_p=f^{\ast}(w.v)_p$,
2. $f^{\ast}((\lambda u+\mu v).w)_p=\lambda f^{\ast}(u.w)_p+\mu f^{\ast}(v.w)_p$.

**Teorema T9.3:** Uma aplicação suave $f:\mathcal S_1\to\mathcal S_2$ entre superfícies é uma isometria local se e só se, para todo o $p\in\mathcal S_1$ e para todo $v,w\in T_p\mathcal S_1$,
$$
f^{\ast}(v.w)_p=v.w\,.
$$

**Proposição T9.4:** Seja $f:\mathcal S_1\to\mathcal S_2$ uma aplicação suave entre superfícies e $p\in\mathcal S_1$.
1. Se $f$ é uma isometria local então $D_pf$ é uma isometria (linear) para todo o $p\in\mathcal S_1$.
2. Se $f$ é uma isometria local então $f$ é um difeomorfismo local.

**Proposição T9.5:** Um difeomorfismo local $f:\mathcal S_1\to\mathcal S_2$ é uma isometria local se e só se para toda a parametrização de um aberto de $\mathcal S_1$, $\sigma$ e $f\circ \sigma$ têm a mesma primeira forma fundamental.

**Definição T9.7:** Seja $\mathcal S$ uma superfície orientada, $\mathcal N:\mathcal S\to \mathbb R^3$ a aplicação suave tal que, para todo o $p\in S$, $\mathcal N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$ e $S^2=\{(x,y,z)\in\mathbb R^3:x^2+y^2+z^2=1\}$

2. Dado $p\in\mathcal S$, definimos a *aplicação de Weingarten*, $\mathcal W_{p,\mathcal S}$, como 
$$
-D_p\mathcal G:T_p\mathcal S\to T_{\mathcal N(p)}S^2.
$$

**Nota:** Para qualquer ponto da esfera $\widetilde{p}$, o plano tangente a $\widetilde{p}$ é perpendicular ao vetor $O\widetilde{p}$, sendo $O$ a origem de $\mathbb R^3$. Dada a unicidade da dualidade entre plano e direcção perpendicular, podemos interpretar $-D_p\mathcal G$ como  $-D_p\mathcal G:T_p\mathcal S\to T_p\mathcal S$.

**Definição T9.9:** Seja $\mathcal S$ uma superfície orientada e $p\in\mathcal S$. Definimos a *segunda forma fundamental de $\mathcal S$ em $p$* como a forma bilinear 
que a $v,w\in T_p\mathcal S$ faz corresponder
$$
\textrm{II}(v,w)_{p,\mathcal S}=\mathcal W_{p,\mathcal S}(v).w,
$$
onde $.$ é a primeira forma fundamental de $p$ em $\mathcal S$.

**Proposição T9.10:** Seja $\mathcal S$ uma superfície orientada com a normal unitária standard $\mathcal N$, $p\in\mathcal S$ e $\sigma(t,s)$ uma parametrização de um aberto de $\mathcal S$ que contém $p$. Então, dados $v,w\in T_p\mathcal S$,
$$
\textrm{II}(v,w)_{p,\mathcal S}=L\,dt(v)dt(w)+M(dt(v)ds(w)+dt(w)ds(v))+N\,ds(v)ds(w),
$$
onde
$$
L=-\partial_t\mathcal N.\partial_t\sigma=\mathcal N.\partial_{tt}\sigma,\,M=-\partial_t\mathcal N.\partial_s\sigma=-\partial_s\mathcal N.\partial_t\sigma=\mathcal N.\partial_{ts}\sigma,\, N=-\partial_s\mathcal N.\partial_s\sigma=\mathcal N.\partial_{ss}\sigma.
$$


**Definição T9.11:** Seja $\mathcal S$ uma superfície orientada segundo a normal unitária standard e $\sigma$ uma parametrização de um aberto de $\mathcal S$. Denominamos
$$
L\,dt^2+2M\,dtds+N\,ds^2,
$$
onde $L,M,N$ são como na Proposição T9.11, por *segunda forma fundamental de $\sigma$*.

**Exercício T9.12:** Calcule a segunda forma fundamental de
$$
\sigma_(t,s)=(t,s,t^2+s^2), (t,s)\in\mathbb R^2.
$$

```python
from sympy import *
t, s = symbols('t s')
g=Matrix([t,s,t**2+s**2]).T
gt=g.diff(t)
gs=g.diff(s)
n=gt.cross(gs)
nus=n/sqrt(n.dot(n))
nus
```

```python
gtt=gt.diff(t)
gts=gt.diff(s)
gss=gs.diff(s)
```

```python
L=gtt.dot(nus)
L
```

```python
M=gts.dot(nus)
M
```

```python
N=gss.dot(nus)
N
```
