**Nota:** Seja $\mathcal S$ uma superfície orientada, $\mathcal N:\mathcal S\to \mathbb R^3$ uma aplicação suave tal que, para todo o $p\in S$, $\mathcal N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$ e $\gamma:I\to \mathcal S$ uma curva de velocidade unitária. Como $\dot{\gamma}\in T_{\gamma(t)}\mathcal S$, $\dot{\gamma}(t)$ e $\mathcal N(\gamma(t))$ são perpendiculares. Os vectores $\dot{\gamma}(t)$, $\mathcal N(\gamma(t))$ e $\mathcal N(\gamma(t))\times \dot{\gamma(t)}$ são perpendiculares. Para simplificar a notação vamos omitir pontos. Dado que $\gamma$ tem velocidade unitária, $\dot{\gamma}$ e $\ddot{\gamma}$ são perpendiculares, logo pertence ao complemento ortogonal de $\dot{\gamma}$, pelo que será combinação linear de $\mathcal N$ e $\mathcal N\times \dot{\gamma}$. Seja
$$
\ddot{\gamma}=\kappa_n\, \mathcal N+\kappa_g\, \mathcal N\times \dot{\gamma}.
$$

**Definição T10.1:** Seja $\mathcal S$ uma superfície orientada, $\mathcal N:\mathcal S\to \mathbb R^3$ uma aplicação suave tal que, para todo o $p\in S$, $\mathcal N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$ e $\gamma:I\to \mathcal S$ uma curva de velocidade unitária. Os escalares $\kappa_n$ e $\kappa_g$, da combinação linear indicada na nota anterior, são denominados por, respectivamente, *curvatura normal* e *curvatura geodésica* de $\gamma$.

**Nota:** O sinal de $\kappa_n$ e $\kappa_g$ depende da escolha do sentido de $\mathcal N$.

**Proposição T10.2:** Seja $\mathcal S$ uma superfície orientada, $\mathcal N:\mathcal S\to \mathbb R^3$ uma aplicação suave tal que, para todo o $p\in S$, $\mathcal N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$ e $\gamma:I\to \mathcal S$ uma curva de velocidade unitária. Então
$$
\kappa_n=\ddot{\gamma}.\,\mathcal N,\,\kappa_g=\ddot{\gamma}.(\mathcal N\times \dot{\gamma}),\, \kappa^2=\kappa_n^2+\kappa_g^2,\, \kappa_n=\kappa\, \textrm{cos}(\theta),\, \kappa_g=\kappa\,\textrm{sen}(\theta) \textrm{ ou } \kappa_g=-\kappa\,\textrm{sen}(\theta),
$$
onde $\kappa$ é a curvatura de $\gamma$ e $\theta$ é o ângulo entre $\mathcal N$ e a normal principal de $\gamma$.

**Proposição T10.3:** Seja $\mathcal S$ uma superfície orientada, $\gamma:I\to \mathcal S$ uma curva de velocidade unitária. Então
$$
\kappa_n=\textrm{II}(\dot{\gamma}(t),\dot{\gamma}(t))_{\gamma(t),\mathcal S}.
$$
Além disso, se $\sigma$ é uma parametrização de um aberto de $\mathcal S$ que contém a curva $\gamma(t)=\sigma(u(t),v(t))$, temos que
$$
\kappa_n=L\dot{u}^2+2M\dot{u}\dot{v}+N\dot{v}^2.
$$

**Nota:** Seja $\gamma:I\to \mathcal S$ uma curva regular e $\gamma_1,\gamma_2$ duas reparametrizações com velocidade unitária de $\gamma$. Então a curvatura normal de $\gamma_1$ e $\gamma_2$ é igual (porquê?) e o módulo da curvatura geodésica é igual (porquê?).

**Definição T10.4:** Seja $\mathcal S$ uma superfície orientada e $\gamma:I\to \mathcal S$ uma curva regular. Definimos curvatura normal e curvatura geodésica de $\gamma$ como a curvatura normal e curvatura geodésica de uma reparametrização com velocidade unitária de $\gamma$.

**Nota:** A definição de curvatura geodésica de uma curva que não seja de velocidade unitária é única a menos de um sinal.

**Definição T10.5:** Seja $\mathcal S$ uma superfície e $\gamma:I\to \mathcal S$ uma curva regular. Denominamos por *campo vetorial ao longo de $\gamma$* a uma aplicação do tipo $v:I\to \mathbb R^3$ tal que, para todo o $t\in I$, $v(t)\in T_{\gamma(t)}\mathcal S$. Interpretamos as imagens $v(t)$ como vetores.

**Definição T10.6:** Seja $\mathcal S$ uma superfície orientada, $\gamma:I\to S$ uma curva e $v:I\to \mathbb R^3$ um campo vetorial ao longo de $\gamma$. Definimos $\nabla_{\gamma}v(t)$, a *derivada covariante de $v$ ao longo de $\gamma$*, como a projecção ortogonal de $\dot{v}(t)$ em $T_{\gamma(t)}S$, ou seja,
$$
\nabla_{\gamma}v(t)=\dot{v}(t)-(\dot{v}(t)\,.\,\mathcal N(t))\mathcal N(t),
$$
 onde $\mathcal N:\mathcal S\to \mathbb R^3$ é uma aplicação suave tal que, para todo o $p\in S$, $\mathcal N(p)$ é um vetor unitário perpendicular a $T_p\mathcal S$.

**Nota:** A definição T10.6 depende apenas da construção do complemento ortogonal de espaço tangente, pelo que pode ser generalizada ao caso de superfícies não orientadas.

**Definição T10.7:** Seja $\mathcal S$ uma superfície orientada, $\gamma:I\to \mathcal S$ uma curva e $v:\gamma(I)\to \mathbb R^3$ um campo vetorial ao longo de $\gamma$. Dizemos que *$v$ é paralelo ao longo de $\gamma$* se $\nabla_{\gamma}v(t)=0$ para todo o $t\in I$.

**Nota:** Por definição, o campo vetorial $v$ é paralelo ao longo de $\gamma$ se e só se $\dot{v}(t)=(\dot{v}(t)\,.\,\mathcal N(t))\mathcal N(t)$, ou seja, se e só se, para todo o $t\in I$, $\dot{v}(t)$ é perpendicular a $T_{\gamma(t)}\mathcal S$.

**Proposição T10.8 (Equações de Gauss):** Seja $\mathcal S$ uma superfície orientada com a normal unitária standard $\mathcal N$. Seja ainda $\sigma(t,s)$ uma parametrização de um aberto de $\mathcal S$, com primeira forma fundamental
$$
E\,dt^2+2F\,dt\,ds+G\,ds^2
$$
e segunda forma fundamental
$$
L\,dt^2+2M\,dt\,ds+N\,ds^2.
$$
Então
$$
\partial_{tt}\sigma=\Gamma^{1}_{11}\partial_t\sigma+\Gamma^{2}_{11}\partial_s\sigma+ L\,\mathcal N,
$$
$$
\partial_{ts}\sigma=\Gamma^{1}_{12}\partial_t\sigma+\Gamma^{2}_{12}\partial_s\sigma+ M\,\mathcal N,
$$
$$
\partial_{ss}\sigma=\Gamma^{1}_{22}\partial_t\sigma+\Gamma^{2}_{22}\partial_s\sigma+ N\,\mathcal N
$$
onde
$$
\Gamma^{1}_{11}=\frac{G\partial_t E-2F\partial_t F+F\partial_s E}{2(E\,G-F^2)},\,\Gamma^{2}_{11}=\frac{2E\partial_t F-E\partial_s E-F\partial_t E}{2(E\,G-F^2)},
$$
$$
\Gamma^{1}_{12}=\frac{G\partial_s E-F\partial_t G}{2(E\,G-F^2)},\,\Gamma^{2}_{12}=\frac{E\partial_t G-F\partial_s E}{2(E\,G-F^2)},
$$
$$
\Gamma^{1}_{22}=\frac{2G\partial_s F-G\partial_t G-F\partial_s G}{2(E\,G-F^2)},\,\Gamma^{2}_{22}=\frac{E\partial_s G-2F\partial_s F+F\partial_t G}{2(E\,G-F^2)}.
$$
Denominamos os coeficientes $\Gamma^{1}_{11},\Gamma^{2}_{11},\Gamma^{1}_{12},\Gamma^{2}_{12},\Gamma^{1}_{22},\Gamma^{2}_{22}$ por *símbolos de Christoffel*.

**Proposição T10.9:** Seja $\sigma(u,w)$ uma parametrização de um aberto de uma superfície orientada $\mathcal S$, $\gamma(t)=\sigma(u(t),w(t))$, $t\in I$, uma curva regular e
$$
v(t)=\alpha(t)\partial_u\sigma(u(t),w(t))+\beta(t)\partial_w\sigma(u(t),w(t)),\, t\in I
$$
um campo vetorial ao longo de $\gamma$. Temos que $v$ é paralelo ao longo de $\gamma$ se e só se as seguintes equações são verificadas em todos os pontos de $I$:
$$
\dot{\alpha}+\left(\Gamma^{1}_{11}\dot{u}+\Gamma^{1}_{12}\dot{w}\right)\alpha+\left(\Gamma^{1}_{12}\dot{u}+\Gamma^{1}_{22}\dot{w}\right)\beta=0,
$$
$$
\dot{\beta}+\left(\Gamma^{2}_{11}\dot{u}+\Gamma^{2}_{12}\dot{w}\right)\alpha+\left(\Gamma^{2}_{12}\dot{u}+\Gamma^{2}_{22}\dot{w}\right)\beta=0.
$$

**Nota:** AS equações da proposição T10.9 formam um sistema de equações de diferenciais, para as quais existe solução única se dada uma condição inicial do tipo

$$
onde $t_0\in I$, $\alpha_0,\beta_0\in\mathbb R$ e $v(t_0)\in T_{\gamma(t_0)}\mathcal S$.
