
**Proposição T13.1 (Equações de Codazzi-Mainardi):** Sejam $Edu^2 + 2Fdudv + Gdv^2$ e $Ldu^2 + 2Mdudv + Ndv^2$ as primeira e segunda formas fundamentais de uma parametrização $\sigma(u, v)$. Então:
$$\partial_v L - \partial_u M = L\Gamma^1_{12} + M(\Gamma^2_{12} - \Gamma^1_{11}) - N\Gamma^2_{11},$$
$$\partial_v M - \partial_u N = L\Gamma^1_{22} + M(\Gamma^2_{22} - \Gamma^1_{12}) - N\Gamma^2_{12}.$$

**Proposição T13.2 (Equações de Gauss):** Sejam $Edu^2 + 2Fdudv + Gdv^2$ e $Ldu^2 + 2Mdudv + Ndv^2$ as primeira e segunda formas fundamentais de uma parametrização $\sigma(u, v)$. Então:

$$E\mathcal K = \partial_v(\Gamma^2_{11}) - \partial_u(\Gamma^2_{12}) + \Gamma^1_{11}\Gamma^2_{12} + \Gamma^2_{11}\Gamma^2_{22} - \Gamma^1_{12}\Gamma^2_{11} - (\Gamma^2_{12})^2,$$
$$F\mathcal K = \partial_u(\Gamma^1_{12})- \partial_v(\Gamma^1_{11}) + \Gamma^2_{12}\Gamma^1_{12} - \Gamma^2_{11}\Gamma^1_{22},$$
$$F\mathcal K = \partial_v(\Gamma^2_{12}) - \partial_u(\Gamma^2_{22}) + \Gamma^1_{12}\Gamma^2_{12} - \Gamma^1_{22}\Gamma^2_{11},$$
$$G\mathcal K = \partial_u(\Gamma^1_{22}) -\partial_v(\Gamma^1_{12}) + \Gamma^1_{22}\Gamma^1_{11} + \Gamma^2_{22}\Gamma^1_{12} - (\Gamma^1_{12})^2 - \Gamma^2_{12}\Gamma^1_{22},$$
onde $\mathcal K$ é a curvatura Gaussiana da superfície.


**Teorema 13.3:**

$i)$ Sejam $\sigma:U\to\mathbb R^3$ and $\widetilde \sigma:U\to\mathbb R^3$ duas parametrizações com a mesma primeira e segunda formas fundamentais. Então existe uma isometria directa $M$ de $\mathbb R^3$ tal que $\widetilde \sigma=M\circ\sigma$.

$ii)$ Seja $V$ um aberto de $\mathbb R^2$ e sejam $E, F, G, L, M, N$ funções suaves em $V$, tais que $E,F,EG-F^2>0$ e satisfizem as equações de Gauss e de Codazzi–Mainardi. Então, dado $(u_0,v_0)\in V$, existe um aberto $U$ de $\mathbb R^2$, contido em $V$ e que contém $(u_0,v_0)$, e uma parametrização de um aberto de uma superfície $\sigma:U\to\mathbb R^3$ com primeira e segunda forma, respectivamente, $Edu^2 + 2Fdudv + Gdv^2$ e $Ldu^2 + 2Mdudv + Ndv^2$.

**Exemplo T13.4:** Calcule, a menos de uma isometria directa, a superfície com primeira e segunda forma fundamentais, respectivamente, $du^2 + dv^2$ e $-du^2$.

**Teorema Egregium de Gauss:** Seja $f:\mathcal S_1\to \mathcal S_2$ uma isometria local entre superfícies orientadas, com curvatura Gaussiama, respectivamente $\mathcal K_1$ e $\mathcal K_2$. Então
$$
\mathcal K_1=\mathcal K_2\circ f.
$$

**Corolário T13.5:** Dada uma parametrização $\sigma(u,v)$ de um aberto de uma superfície, temos
$$\mathcal K = \frac{\begin{vmatrix} -\frac{1}{2}\partial_{vv}E + \partial_{uv}F - \frac{1}{2}\partial_{uu}G & \frac{1}{2}\partial_u E& \partial_u F- \frac{1}{2}\partial_v E \\ \partial_v F- \frac{1}{2}\partial_u G & E & F \\ \frac{1}{2}\partial_v G & F & G \end{vmatrix} - \begin{vmatrix} 0 & \frac{1}{2}\partial_v E& \frac{1}{2}\partial_u G\\ \frac{1}{2}\partial_v E& E & F \\ \frac{1}{2}\partial_u G & F & G \end{vmatrix}}{(EG - F^2)^2}$$

**Corolário T13.6:** Seja $\sigma(u,v)$ a parametrização de um aberto de uma superfície.

$i)$ Se $F = 0$, então
$$\mathcal K = -\frac{1}{2\sqrt{EG}} \left\{ \frac{\partial}{\partial u} \left( \frac{\partial_u G}{\sqrt{EG}} \right) + \frac{\partial}{\partial v} \left( \frac{\partial_v E}{\sqrt{EG}} \right) \right\}.$$

$ii)$ Se $E = 1$ e $F = 0$, então
$$\mathcal K = -\frac{1}{\sqrt{G}} \frac{\partial^2 \sqrt{G}}{\partial u^2}.$$

$$
onde $f>0$ e $\dot{f}^2+\dot{g}^2=1$.
