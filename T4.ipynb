**Definição T4.1:** Seja $\gamma$ uma curva em $\mathbb R^3$ com velocidade unitária e curvatura não nula. Definimos $T$ como o vetor tangente unitário de $\gamma$, ou seja, $T=\dot{\gamma}$. Definimos a *normal principal* de $\gamma$ no ponto $\gamma(t)$, que representamos por $N(t)$, como
$$
\frac{1}{\kappa(t)}\dot{T}(t).
$$
Definimos o vetor *binormal* de $\gamma$ no ponto $\gamma(t)$, que representamos por $B(t)$, como $T(t)\times N(t)$.

**Nota (prove estas afirmações!):** 
* $N(t)$ é um vetor unitário, ou seja $\|N(t)\|=1$.
* Os vetores $T(t)$ e $N(t)$ são são perpendiculares e como tal $B(t)$ é perpendicular a $T(t)$ e $N(t)$.
* O conjunto $\{T(t),N(t),B(t)\}$ é uma base ortonormada de $\mathbb R^3$ tal que
$$
B(t)=T(t)\times N(t),\, N(t)=B(t)\times T(t),\, T(t)=N(t)\times B(t).
$$
* $\dot{B}(t)$ é paralelo a $N(t)$.

**Definição T4.2:** Seja $\gamma$ uma curva em $\mathbb R^3$ com velocidade unitária e curvatura não nula. Definimos *torção* de $\gamma$ no ponto $\gamma(t)$ ao escalar $\tau(t)$ tal que $\dot{B}(t)=-\tau(t)N(t)$.

**Definição T4.4:** Seja $\gamma$ uma curva em $\mathbb R^3$ com curvatura não nula. Definimos *torção* de $\gamma$ no ponto $\gamma(t)$ como a torção de uma sua reparametrização de velocidade unitária.

**Proposição T4.5:** Seja $\gamma$ uma curva em $\mathbb R^3$ com curvatura não nula. Então
$$
\tau(t)=\frac{(\dot{\gamma}(t)\times \ddot{\gamma}(t))\cdot \frac{d^3\gamma}{d t^3}(t)}{\|\dot{\gamma}(t)\times \ddot{\gamma}(t)\|^2}.
$$

**Exemplo T4.6:** Calcule a torção da curva do exemplo T3.3.

```python
from sympy import *
t, a, b = symbols('t a b')
g=Matrix([a*cos(t),a*sin(t),b*t]).T
dg=g.diff(t)
d2g=g.diff(t,2)
d3g=g.diff(t,3)
d3g
```

```python
c1=dg.cross(d2g)
simplify(c1.dot(d3g)/c1.dot(c1))
```
